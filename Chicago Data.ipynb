# SQL Practice

Below are the use of my SQL answers for the following questions from the following course `Databases and SQL For Data Science With Python` on Coursera. I have been able to properly use DML/DDL SQL statements, string patterns, grouping, and work through functions and sub-queries through this practice.

```python
%%sql

# Find the total number of crimes recorded in the CRIME table.
# COUNT(*) returns the number of rows in a specified table.

SELECT COUNT(*) FROM chicago_crime_data
```

```python
%%sql

# List community areas with per capita income less than 11000
# WHERE statement = filtering; fulfills a specific condition.

SELECT community_area_name
FROM chicago_census_data
WHERE per_capita_income < 11000
```

```python
%%sql

# List all case numbers for crimes involving minors.
# WHERE statement used again, LIKE = specified pattern
# % specifies multiple characters

SELECT case_number
FROM chicago_crime_data
WHERE description LIKE '%MINOR%'
```

```python
%%sql

# List all kidnapping crimes involving a child.

SELECT * FROM chicago_crime_data
WHERE primary_type LIKE '%KIDNAP%' AND description LIKE '%CHILD%'
```

```python
%%sql

# What kinds of crimes were recorded at schools?

SELECT DISTINCT(primary_type)
FROM chicago_crime_data
WHERE location_description LIKE '%SCHOOL%'
```

```python
%%sql

# List the average safety score for all types of schools.
# AVG function = calculates the average
# GROUP BY = same values into summary

SELECT "Elementary, Middle, or High School", AVG(safety_score)
FROM chicago_schools
GROUP BY "Elementary, Middle or High School"
```

```python
%%sql

# List 5 community areas with highest % of households below poverty line
# ORDER BY = sorts in ascending or descending
# DESC = descending, LIMIT = limiting the number of rows

SELECT community_area_name, percent_households_below_poverty
FROM census_data
ORDER BY percent_households_below_poverty DESC LIMIT 5
```

```python
%%sql

# Which community area is most crime prone?

SELECT community_area_number, COUNT(*) AS frequency FROM chicago_crime_data
GROUP BY community_area_number ORDER BY frequency DESC LIMIT 1
```

```python
%%sql

# Use a sub-query to find the name of the community area with the highest hardship index.
# Note the parantheses, I want to SELECT the MAX (which the largest value) of hardship_index

SELECT community_area_name
FROM chicago_census_data
WHERE hardship_index = (SELECT MAX(hardship_index) FROM chicago_census_data)
```
